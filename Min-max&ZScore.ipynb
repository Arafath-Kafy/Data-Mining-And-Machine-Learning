Z-score help to remove outliers from dataset

```python
from scipy.stats import norm
import numpy as np
plt.hist(df['Length'], bins=15, rwidth=0.8, density=True)
plt.xlabel("price")
plt.ylabel("count")
rng = np.arange(df.Length.min(),df.Length.max(),0.05)
plt.plot(rng,norm.pdf(rng,df.Length.mean(),df.Length.std()))
```

```python
new_df = df
new_df
```

```python
#zscore for weight column
new_df['zscore'] = (new_df.Weight	 - new_df.Weight.mean())/new_df.Weight.std()
new_df
```

```python
#shows outliers
new_df[(new_df['zscore'] > 2) | (new_df['zscore'] < -2)]
```
