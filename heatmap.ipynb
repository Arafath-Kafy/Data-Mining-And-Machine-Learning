```python
y1 = d_test.iloc[:,7:8]
y1
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
x_train, x_test, y_train, y_test = train_test_split(x1,y1,random_state = 43, test_size=0.2)
```

```python
reg = LinearRegression()
reg = reg.fit(x_train, y_train)
reg.score(x_test, y_test)
```

```python
x_train_pred = reg.predict(x_train)
x_test_pred = reg.predict(x_test)
```

```python
#red is the best fit line
plt.scatter(x_train, x_train_pred, color="blue")
plt.plot(x_test, x_test_pred, color="red")
plt.title("Income vs House Price (Training Dataset)")
plt.xlabel("House Price")
plt.ylabel("Income")
plt.show()
```
